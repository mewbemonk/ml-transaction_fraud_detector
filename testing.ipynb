```python
df = pd.get_dummies(df,columns=['type'])
df.head()
```

```python
col = ['type_CASH_IN', 'type_CASH_OUT','type_DEBIT', 'type_PAYMENT', 'type_TRANSFER']

for i in col:
    df[i] = df[i].astype(int)


df.head()
```

```python
df = df.drop_duplicates()

df.duplicated().sum()
```

```python
df['balance_diff_orig'] = df['oldbalanceOrg']-df['newbalanceOrig']
df['balance_diff_dest'] = df['oldbalanceDest']-df['newbalanceDest']
df
```

```python
df['isFraud'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler

numeric_col = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest','newbalanceDest']

scaler = StandardScaler()

df[numeric_col] = scaler.fit_transform(df[numeric_col])

df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline



x = df.drop('isFraud',axis=1)
y = df['isFraud']

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier


models = {

    "LogisticRegression":{
        "model": LogisticRegression(),
        "params":{
            "C":[0.01,0.1,1,10],
            "max_iter":[1000]
        }
    },

    "KNN":{
        "model": KNeighborsClassifier(),
        "params":{
            "n_neighbors":[3,5,7,9],
            "weights":["uniform","distance"]
        }
    },

    "SVM":{
        "model": SVC(),
        "params":{
            "C":[0.1,1,10],
            "kernel":["linear","rbf"]
        }
    },

    "NaiveBayes":{
        "model": GaussianNB(),
        "params":{
            "var_smoothing":[1e-9,1e-8,1e-7]
        }
    },

    "DecisionTree":{
        "model": DecisionTreeClassifier(),
        "params":{
            "max_depth":[5,10,20,None],
            "criterion":["gini","entropy"]
        }
    },

    "RandomForest":{
        "model": RandomForestClassifier(),
        "params":{
            "n_estimators":[100,200],
            "max_depth":[10,20,None]
        }
    },

    "XGBoost":{
        "model": XGBClassifier(),
        "params":{
            "n_estimators":[100,200],
            "learning_rate":[0.01,0.1],
            "max_depth":[3,6]
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV




res = []

for name,p in models.items():


    cv = GridSearchCV(p['model'],param_grid=p['params'],cv=5)

    cv.fit(x_train,y_train)
    
    res.append({
        "model":name,
        "cv-score":cv.best_score_,
        "param": cv.best_params_,
        "estimator": cv.best_estimator_
    })


res
```

```python
import matplotlib.pyplot as plt

models = [i["model"] for i in res]
scores = [i["cv-score"] for i in res]

plt.figure(figsize=(8,5))

plt.bar(models, scores)

plt.xlabel("Models")
plt.ylabel("CV Score")
plt.title("Model Comparison (Cross Validation Score)")

plt.xticks(rotation=45) 

plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

best_model = RandomForestClassifier(
    n_estimators = 100,
    max_depth = None,
    max_features ='sqrt',
    min_samples_split = 2,
    )
```

```python
best_model.fit(x_train,y_train)

y_pred = best_model.predict(x_test)

accuracy = classification_report(y_test,y_pred)

print(accuracy)
```

```python
from sklearn.model_selection import cross_val_score



scores = cross_val_score(best_model, x, y, cv=5)

print("Cross validation scores:", scores)
print("Mean CV score:", scores.mean())
```
